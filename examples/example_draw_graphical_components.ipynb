```python
from PyCarPlt.lib_graphic import *

radius = 1

fig, ax = plt.subplots(1,1)
circle = plt.Circle((0, 0), radius,facecolor=(1,1,1,0),edgecolor="black",linewidth=0.5)
ax.add_patch(circle)
ax.plot(0,0,'k')
arc_sector((0,0), radius, 90, 180,ax=ax, fill=True, color='black')
arc_sector((0,0), radius, 270, 360,ax=ax, fill=True, color='black')
ax.set_aspect('equal', adjustable='box')
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(0,0,'k')
arc_arrow((0,0), 0.2, 0, 180, ax=ax,linewidth=0.5)
ax.set_aspect('equal', adjustable='box')
```
