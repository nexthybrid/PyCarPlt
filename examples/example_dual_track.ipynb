Load data from csv and plot the pose

```python
# read from the csv file 'drift_data.csv' using pandas
import pandas as pd
df = pd.read_csv('examples/drift_data.csv')

# use a specified row from the csv file
row_num = 1000
print(df.iloc[row_num])
psi = df.iloc[row_num]['psi']
xdot = df.iloc[row_num]['xdot']
ydot = df.iloc[row_num]['ydot']
F_tire_fl = [df.iloc[row_num]['Fx_0'], df.iloc[row_num]['Fy_0']]
F_tire_fr = [df.iloc[row_num]['Fx_1'], df.iloc[row_num]['Fy_1']]
F_tire_rl = [df.iloc[row_num]['Fx_2'], df.iloc[row_num]['Fy_2']]
F_tire_rr = [df.iloc[row_num]['Fx_3'], df.iloc[row_num]['Fy_3']]
slip_angle_fl = df.iloc[row_num]['Alpha_0']
slip_angle_fr = df.iloc[row_num]['Alpha_1']
slip_angle_rl = df.iloc[row_num]['Alpha_2']
slip_angle_rr = df.iloc[row_num]['Alpha_3']

v.update_body_pose(heading=psi,vel_bframe=(xdot,ydot))
v.update_tire_forces(tire_forces=(F_tire_fl,F_tire_fr,F_tire_rl,F_tire_rr),\
                     slip_angles=(slip_angle_fl,slip_angle_fr,slip_angle_rl,slip_angle_rr))
v.draw_vehicle()
```
